```python
# K MEANS CLUSTERING

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/Mall_Customers.csv')
df.head(5)

# X = df.drop(['CustomerID','Genre','Age'],axis=1)
X = df.iloc[:,[3,4]].values  # .values converts it into NumPy array
# print(X)

# using elbow method
from sklearn.cluster import KMeans
wcss = []          # You try different cluster counts (1 → 10) Lower WCSS = better clustering
for i in range(1,11):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)  # As k increases, WCSS decreases But after a point → improvement slows down hat point = Elbow Point (Optimal K)
# Inertia = total squared distance of all points from their cluster centers
# kmeans.inertia_ - It gives WCSS (Within Cluster Sum of Squares)

plt.plot(range(1,11),wcss)
plt.title("the elbow method")
plt.xlabel("clusters")
plt.ylabel("wcss")
plt.show()

# applying the k means to the dataset
kmeans = KMeans(n_clusters=6, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# visualising the clusters
# first zero - Selects all data points that belong to cluster 0 ----left = cluster , right = income,spending score
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=100,c='red',label='cluster 1')  # X[:,0] → income X[:,1] → spending score
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=100,c='yellow',label='cluster 2')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=100,c='green',label='cluster 3')
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],s=100,c='blue',label='cluster 4')
plt.scatter(X[y_kmeans==4,0],X[y_kmeans==4,1],s=100,c='cyan',label='cluster 5')
plt.scatter(X[y_kmeans==5,0],X[y_kmeans==5,1],s=100,c='purple',label='cluster 6')
# plt.scatter(X[y_kmeans==6,0],X[y_kmeans==6,1],s=100,c='pink',label='cluster 7')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='black',label='centroid')
plt.title("cluster of client")
plt.xlabel("annual income")
plt.ylabel("spending score")
plt.show()
```

```python
# HIERARCHICAL CLUSTERING

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/Mall_Customers.csv')
df.head(5)

# X = df.drop(['CustomerID','Genre','Age'],axis=1)
X = df.iloc[:,[3,4]].values  # .values converts it into NumPy array

import scipy.cluster.hierarchy as sch
dg = sch.dendrogram(sch.linkage(X,method='ward'))

plt.title("DENDROGRAM")
plt.xlabel("CUSTOMERS")
plt.ylabel("EUCLIDIAN DISTANCE")
plt.show()

from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage='ward') #Ward method : minimizes variance within clusters
y_hc = ac.fit_predict(X)

plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=100,c='red',label='cluster 1')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=100,c='yellow',label='cluster 2')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=100,c='green',label='cluster 3')
plt.scatter(X[y_hc==3,0],X[y_hc==3,1],s=100,c='blue',label='cluster 4')
plt.scatter(X[y_hc==4,0],X[y_hc==4,1],s=100,c='cyan',label='cluster 5')

plt.title("cluster of client")
plt.xlabel("annual income")
plt.ylabel("spending score")
plt.show()
```

```python
# HIERARCHICAL CLUSTERING

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/Mall_Customers.csv')
df.head(5)

# X = df.drop(['CustomerID','Genre','Age'],axis=1)
X = df.iloc[:,[3,4]].values  # .values converts it into NumPy array

import scipy.cluster.hierarchy as sch
dg = sch.dendrogram(sch.linkage(X,method='ward'))

plt.title("DENDROGRAM")
plt.xlabel("CUSTOMERS")
plt.ylabel("EUCLIDIAN DISTANCE")
plt.show()

from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5,metric='manhattan',linkage='complete') # Distance between two clusters = maximum distance between any two points
y_hc = ac.fit_predict(X)

plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=100,c='red',label='cluster 1')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=100,c='yellow',label='cluster 2')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=100,c='green',label='cluster 3')
plt.scatter(X[y_hc==3,0],X[y_hc==3,1],s=100,c='blue',label='cluster 4')
plt.scatter(X[y_hc==4,0],X[y_hc==4,1],s=100,c='cyan',label='cluster 5')

plt.title("cluster of client")
plt.xlabel("annual income")
plt.ylabel("spending score")
plt.show()
```

```python
# MEAN SHIFT CLUSTERING

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift

df = pd.read_csv('/content/sample_data/Mall_Customers.csv')
df.head(5)

X = df.iloc[:,[3,4]].values
ms = MeanShift(bandwidth=20)
ms.fit(X)
labels = ms.labels_
cluster_centers = ms.cluster_centers_
print(labels)

n_clusters = len(np.unique(labels))

colors = ["red","green","blue","yellow","cyan","pink","purple"]

for i in range(n_clusters):
  cluster = X[labels==i]
  plt.scatter(cluster[:,0],cluster[:,1],s=100,c=colors[i],label=f'cluster {i+1}')

plt.title("cluster of client")
plt.xlabel("annual income")
plt.ylabel("spending score")
plt.legend()
plt.show()
```
