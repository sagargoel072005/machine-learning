```python
# map(function,iterable)
import math
radii = [2,3,4,5,6]
radii2 = [12,13,14,15,16]
area = lambda x : math.pi*(x**2)
areas1 = []
for r in radii:
   a = area(r)
   areas1.append(a)
print(areas1)
areas = list(map(area, radii2))
print(areas)
```

```python
# map(function,iterable)
temp = [("Mumbai",35),("Delhi",45),("Banglore",55),("Pune",65),("Goa",75)]
cel_to_f = lambda x : (x[0],(9/5)*x[1]+32)
ans = list(map(cel_to_f,temp))
print(ans)
```

```python
# filter(function,data)
import statistics
data = [11,29,32,48,53,67,74,86,95,10]
ans = statistics.mean(data)
print(ans)
ans2 = list(filter(lambda x : x>ans , data))
print(ans2)
```

```python
# reduce(function,data)
from functools import reduce
data = [12,23,34,45,56,67,78,89,90]
multiplier = lambda x,y : x*y
print(reduce(multiplier,data))
adder = lambda x,y : x+y
print(reduce(adder,data))
```

```python
# file handling - read operation

with open("myfile.txt", "w") as f:
    f.write("Hello Sagar!\n")
    f.write("This is a text file in Google Colab.")

file = open('myfile.txt','r')
for line in file:
  print(line)

# after loop , The file pointer goes to END of file,thats why we use seek

print("-----------------------------------------------")
file.seek(0)
print(file.read())
print("-----------------------------------------------")
file.seek(0)
print(file.read(15))
print("-----------------------------------------------")
file.seek(0)
print(file.readline())
```

```python
# control structure

def reverse(x):
    ans=0;
    while(x!=0):
        digit = x % 10
        ans = (ans*10) + digit
        x = x//10
    return ans
print(reverse(123))
age=20
if age > 10 :
  print("old")
else:
  print("young")
```

```python
# OOPS IN PYTHON

class Rectangle:
  def __init__(self):  # self parameter refers to the newly created instance of the class
    self.length = 10
    self.breadth = 5
rect = Rectangle()
print(rect.breadth)

class Rectangle1:
  def __init__(self,length,breadth):
    self.length = length
    self.breadth = breadth
rect1 = Rectangle1(45,455)
print(rect1.breadth)
```

```python
# OOPS IN PYTHON

class Circle:
  pi = 3.14 # class variable or constant value
  def __init__(self,radius):
    self.radius = radius  # instance variable
    self.area = Circle.pi*radius*radius
c1 = Circle(7)
print(c1.radius)
print(c1.area)
```

```python
# Inheritance in Python

class Animal:
  def speak(self):
    print("animal makes a sound")
class Dog(Animal):
  def bark(self):
    print("Dog barks")
  # Method Overriding in Python
  def speak(self):   # overriding parent method
        print("Dog barking....")
d = Dog()
d.speak()
d.bark()
```

```python
class Rectangle:
  def __init__(self,length,breadth):
    self.length = length
    self.breadth = breadth
class Area(Rectangle):
  def area(self):
    print("area : ", self.length*self.breadth )
rect1 = Area(45,455)
print(rect1.length)
print(rect1.breadth)
rect1.area()
```
